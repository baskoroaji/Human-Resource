## Data Preparation / Preprocessing

```python
df_dummy = df.dropna(subset=['Attrition']).copy()
```

```python
num_cols = df_dummy.select_dtypes(include=[np.number]).columns.drop(['Attrition','EmployeeId'])
cat_nominal = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']
cat_ordinal = ['Education','EnvironmentSatisfaction','JobInvolvement','JobSatisfaction',
               'PerformanceRating','RelationshipSatisfaction','WorkLifeBalance']
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', MinMaxScaler(), num_cols),
    ('ord', OrdinalEncoder(), cat_ordinal),
    ('nom', OneHotEncoder(drop='first'),cat_nominal)
])
```

```python
X = df_dummy.drop('Attrition', axis=1)
y = df_dummy['Attrition']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_preprocessed, y_train)
```

## Modeling

```python
lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train_resampled, y_train_resampled)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_resampled, y_train_resampled)
```

```python
lr_preds = lr_model.predict(X_test_preprocessed)
rf_preds = rf_model.predict(X_test_preprocessed)
```

## Evaluation

```python
print("Logistic Regression Report")
print(classification_report(y_test, lr_preds))
print(confusion_matrix(y_test, lr_preds))

print("\n Random Forest Report")
print(classification_report(y_test, rf_preds))
print(confusion_matrix(y_test, rf_preds))
```

```python
cm_lr = confusion_matrix(y_test, lr_preds)
plt.figure(figsize=(5,4))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
cm_rf = confusion_matrix(y_test, rf_preds)
plt.figure(figsize=(5,4))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
